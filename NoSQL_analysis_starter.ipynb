```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# 6 Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": "20"}

# Use count_documents to display the number of documents in the result
hygiene_score_20_count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {hygiene_score_20_count}")

# Display the first document in the results using pprint
first_hygiene_score_20 = establishments.find_one(query)
pprint(first_hygiene_score_20)
```

```python
# 7 Convert the result to a Pandas DataFrame
query = {"scores.Hygiene": "20"}
hygiene_score_20_data = list(establishments.find(query))
hygiene_score_20_df = pd.DataFrame(hygiene_score_20_data)

# Display the number of rows in the DataFrame
num_rows = len(hygiene_score_20_df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(hygiene_score_20_df.head(10))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# 8 Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "LocalAuthorityName": "London",
    "RatingValue": {"$gte": 4}
}

# Use count_documents to display the number of documents in the result
london_high_rating_count = establishments.count_documents(query)
print(f"Number of establishments in London with a RatingValue greater than or equal to 4: {london_high_rating_count}")

# Display the first document in the results using pprint
first_london_high_rating = establishments.find_one(query)
print("First establishment in London with a high rating:")
pprint(first_london_high_rating)
```

```python
# 9 Convert the result to a Pandas DataFrame
london_high_rating_data = list(establishments.find(query))
london_high_rating_df = pd.DataFrame(london_high_rating_data)

# Display the number of rows in the DataFrame
num_rows = len(london_high_rating_df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(london_high_rating_df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# 10 Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142  # Latitude of "Penang Flavours"
longitude = 0.08384  # Longitude of "Penang Flavours"

query = {
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
    "RatingValue": 5
}

sort = [("scores.Hygiene", 1)]  # Sort by hygiene score in ascending order

# Find the top 5 establishments meeting the criteria
top_5_establishments = list(establishments.find(query).sort(sort).limit(5))

# Print the results
print("Top 5 establishments with RatingValue of 5, sorted by lowest hygiene score:")
pprint(top_5_establishments)
```

```python
# 11 Convert result to a Pandas DataFrame
import pandas as pd

# Create a DataFrame from the query result
top_5_df = pd.DataFrame(top_5_establishments)

# Display the DataFrame
top_5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# 12 Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Execute the aggregation pipeline
result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of establishments in each Local Authority area with a hygiene score of 0:")
for entry in result[:10]:  # Print the top 10 results
    print(f"{entry['_id']}: {entry['count']}")
```

```python
# 13 Convert the result to a Pandas DataFrame
# Create a DataFrame from the aggregation result
local_authority_hygiene_df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
num_rows = local_authority_hygiene_df.shape[0]
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")

local_authority_hygiene_df.head(10)
```
